```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
ano2019 =pd.read_csv("https://raw.githubusercontent.com/LucasmFari/PROJETO5/main/Ano-2019.csv", sep=';', on_bad_lines = 'skip')
ano2019


ano2020 =pd.read_csv("https://raw.githubusercontent.com/LucasmFari/PROJETO5/main/Ano-2020.csv", sep=';', on_bad_lines = 'skip')
ano2020

ano2021 =pd.read_csv("https://raw.githubusercontent.com/LucasmFari/PROJETO5/main/Ano-2021.csv", sep=';', on_bad_lines = 'skip')
ano2021
```

```python

dados = ano2019.merge(ano2020, how='outer')
dados = dados.merge(ano2021, how='outer')
dados 
dadosbackup= dados.copy()
```

```python
dados = dados[['txNomeParlamentar', 'sgUF', 'sgPartido', 'txtDescricao', 'datEmissao', 'vlrLiquido', 'txtTrecho', 'nuDeputadoId']]
```

```python
dados['datEmissao'] = pd.to_datetime(dados.datEmissao)
dados
```

```python

dados['txtTrecho'] = dados['txtTrecho'].str.upper()

dados['txtTrecho'].replace(" ","", regex=True, inplace=True)
dados
```

```python
dados['datEmissao'].fillna(method="ffill", inplace=True)
dados[dados['txtDescricao'] == 'TELEFONIA']
```

## 1- Como se comportam os gastos com cotas parlamentares ao longo do tempo? Existe alguma tendência de aumento ou redução desse custo?


```python
somames = dados[['datEmissao','vlrLiquido']].groupby(by = [dados['datEmissao'].dt.year, dados['datEmissao'].dt.month]).sum()
somames.index.names = ['ano','mes']
somames = somames.reset_index()
somames.index.names = ['meses']

fig = px.line(somames,y='vlrLiquido',labels={'vlrLiquido': 'total'})

fig.show()

```

## 2- Qual a média de gastos por parlamentar? 

```python
dados[['vlrLiquido','txNomeParlamentar']].groupby('txNomeParlamentar').mean()
```

```python
mediaporpartido = dados[['sgPartido', 'vlrLiquido']].groupby('sgPartido').mean().reset_index()
fig = px.bar(mediaporpartido, x='sgPartido', y='vlrLiquido')
fig.show()


```

```python
dados['ano'] = dadosbackup['numAno']

```

```python
mediaporpartido = dados[['sgPartido', 'vlrLiquido', 'ano']].groupby(['sgPartido','ano'
]).mean().reset_index()
fig = px.bar(mediaporpartido, x='sgPartido', y='vlrLiquido', color='ano')
fig.show()
```

##3- Quais foram os parlamentares que mais e que menos consumiram recursos? 

```python
totalparlamentar=dados[['vlrLiquido','txNomeParlamentar']].groupby('txNomeParlamentar').sum().sort_values(by='vlrLiquido')
totalparlamentar.tail()
```

## 4- Quais são as categorias de despesas mais onerosas dentre os recursos destinados às cotas parlamentares?

```python
totaldespesas=dados[['vlrLiquido','txtDescricao']].groupby('txtDescricao').sum().sort_values(by='vlrLiquido')
totaldespesas.tail()
```

```python
list(totaldespesas.tail().T)
```

```python
totaldespesasano=dados[dados['txtDescricao'].isin(list(totaldespesas.tail().T))][['vlrLiquido','txtDescricao','ano']].groupby(["ano",'txtDescricao']).sum().reset_index().sort_values(by='vlrLiquido')
```

```python
fig = px.bar(totaldespesasano, x='ano', y='vlrLiquido', color='txtDescricao')
fig.show()
```
